```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file, assuming the first column is x and the second is y
df = pd.read_csv('contours.csv', header=None, names=['x', 'y'])

# Plot the points as a scatter plot
plt.scatter(df['x'], df['y'])
plt.title('Scatter Plot of Points')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Plot the points as a line plot, connecting each point to the next
plt.plot(df['x'], df['y'])
plt.title('Line Plot of Points')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
